```python
# Import the necessary libraries
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Import the dataset from the path
dataset = pd.read_csv('Data/insurance.csv')
```

```python
# Separating the x and y values from dataset
x = dataset["x"]
y = dataset["y"]
```

```python
# Splitting the x and y variable into training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=7)
```

```python
# We need to reshape X as our data has single independent feature and the model accepts 2D array
x_train = x_train.values.reshape(-1, 1)
x_test = x_test.values.reshape(-1, 1)
```

```python
# Fitting the model with the training data
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
# Pridicting the values of y
y_train_pred = regressor.predict(x_train)
y_test_pred = regressor.predict(x_test)
```

```python
# Code to visualize train prediction
plt.scatter(x_train, y_train, color='red')
plt.plot(x_train, y_train_pred, color='blue')

# Giving a title for our plot
plt.title('Linear Regression Model (Testing)')
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.show()
```

```python
# Code to visualize test prediction
plt.scatter(x_test, y_test, color='red')
plt.plot(x_test, y_test_pred, color='blue')

# Giving a title for our plot
plt.title('Linear Regression Model (Testing)')
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.show()
```
